```python
print(KNN_df.columns)
print(SVM_df.columns)
print(tree_df.columns)
```

```python

# Agregar columna "modelo" a cada dataframe
KNN_df['modelo'] = 'KNN'
SVM_df['modelo'] = 'SVM'
tree_df['modelo'] = 'DTC'
```

```python
# Seleccionar columnas
KNN_df = KNN_df[['name', 'SMILES', 'Total Molweight', 'cLogP', 'H-Acceptors', 'H-Donors', 'Total Surface Area', 'Relative PSA', 'Rotatable Bonds', 'Predicciones', 'modelo']]
SVM_df = SVM_df[['name', 'SMILES', 'Total Molweight', 'cLogP', 'H-Acceptors', 'H-Donors', 'Total Surface Area', 'Relative PSA', 'Rotatable Bonds', 'Prediction', 'modelo']]
tree_df = tree_df[['name', 'SMILES', 'Total Molweight', 'cLogP', 'H-Acceptors', 'H-Donors', 'Total Surface Area', 'Relative PSA', 'Rotatable Bonds','Prediccion', 'modelo']]

# Cambiar nombre a columnas
KNN_df.columns = ['Name', 'SMILES', 'Total Molweight', 'cLogP', 'H-Acceptors', 'H-Donors', 'Total Surface Area', 'Relative PSA', 'Rotatable Bonds', 'Geroprotectors', 'Modelo']
SVM_df.columns = ['Name', 'SMILES', 'Total Molweight', 'cLogP', 'H-Acceptors', 'H-Donors', 'Total Surface Area', 'Relative PSA', 'Rotatable Bonds', 'Geroprotectors', 'Modelo']
tree_df.columns = ['Name', 'SMILES', 'Total Molweight', 'cLogP', 'H-Acceptors', 'H-Donors', 'Total Surface Area', 'Relative PSA', 'Rotatable Bonds', 'Geroprotectors', 'Modelo']
```

```python
# positivos indicados en Geroprotectors para cada dataframe

# Obtener los positivos para cada modelo
KNN_positivos = KNN_df[KNN_df['Geroprotectors'] == 1]['Name'].tolist()
SVM_positivos = SVM_df[SVM_df['Geroprotectors'] == 1]['Name'].tolist()
tree_positivos = tree_df[tree_df['Geroprotectors'] == 1]['Name'].tolist()

print("KNN Positivos:", len(KNN_positivos))
print("SVM Positivos:", len(SVM_positivos))
print("Árbol Positivos:", len(tree_positivos))
```

```python
# cuantos datos positivos (1) compartes los modelos SVM y tree

# Encontrar la intersección entre SVM y Árbol de decisión
intersection_svm_tree = list(set(SVM_positivos) & set(tree_positivos))

print("Datos positivos compartidos entre SVM y Árbol de decisión:", len(intersection_svm_tree))
```

```python
# cuantos datos positivos (1) compartes los modelos SVM y KNN

# Encontrar la intersección entre SVM y KNN
intersection_svm_knn = list(set(SVM_positivos) & set(KNN_positivos))

print("Datos positivos compartidos entre SVM y KNN:", len(intersection_svm_knn))
```

```python
#  cuantos datos positivos (1) compartes los modelos  tree y KNN

# Encontrar la intersección entre Árbol de decisión y KNN
intersection_tree_knn = list(set(tree_positivos) & set(KNN_positivos))

print("Datos positivos compartidos entre Árbol de decisión y KNN:", len(intersection_tree_knn))
```
